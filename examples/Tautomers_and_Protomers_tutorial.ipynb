Tautomerization and Protomerization Packages

```python
import os
import sys
import pandas as pd
sys.path.append("/work/users/amahamadakalapuwage/SmartCADD_Ayesh/SmartCADD/")

import smartcadd
from smartcadd.modules import Tautomers, Protomers
from smartcadd.data import Compound
from rdkit import Chem
from rdkit.Chem import Draw
```

```python
template_compound_smiles = {
    "example1": "Oc1c(cccc3)c3nc2ccncc12",
    "example2": "CN=c1nc[nH]cc1",
}

template_compounds = [
    Compound(smiles=smiles, id=drug)
    for drug, smiles in template_compound_smiles.items()

]
```

Protomerizationa

```python
protomerization = Protomers()
protomerized_compounds = protomerization.run(template_compounds)
```

```python
print('Example2: Original Compound')
Chem.MolFromSmiles('CN=c1nc[nH]cc1')
```

```python
Draw.MolsToGridImage(protomerized_compounds[1], molsPerRow=3)
```

Tautomerization

```python
tautomerization = Tautomers()
tautomerized_compounds = tautomerization.run(template_compounds)
```

```python
print('Example2: Original Compound')
Chem.MolFromSmiles('CN=c1nc[nH]cc1')
```

```python
Draw.MolsToGridImage(tautomerized_compounds[1], molsPerRow=3)
```
